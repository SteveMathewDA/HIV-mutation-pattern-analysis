```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
df = pd.read_csv("OneHot_top30mutations.csv")
```

```python
df
```

```python
X = df.drop(columns = ['Best drug'])
```

```python
Y = df['Best drug']
```

```python
X
```

```python
Y
```

```python
model = DecisionTreeClassifier()
model.fit(X,Y)
```

```python
tree.export_graphviz(model,out_file = 'drug_recommender_top30.dot',
                    feature_names = ['30N','32I','46I','46L','47A','47V','48A','48S','48T','48Q','48L',
                                     '48M','48V','50L','50V','54A','54T','54S','54L','54M','54V','76V',
                                     '82A','82C','82F','82L','82M','82T','82S','84A','84C','84V','88S',
                                     '88T','88G','90M'],
                    class_names = Y.unique(),rounded = True,filled = True)
```
