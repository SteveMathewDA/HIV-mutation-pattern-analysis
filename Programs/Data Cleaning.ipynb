```python
import pandas as pd
```

```python
df = pd.read_csv("CleanedHIVdataset.csv")
df

mutationlist = ['Mutation 1',
               'Mutation 2',
               'Mutation 3',
               'Mutation 4',
               'Mutation 5',
               'Mutation 6',
               'Mutation 7']

for i in mutationlist:
    df[i] = df[i].fillna(0)
df
```

```python
Unique_Alpha = ['0', 'A', 'C', 'F', 'I', 'L', 'M', 'N', 'Q', 'S', 'T', 'V']
encoding = {}
num = 0
for i in Unique_Alpha:
    encoding[i] = num
    num += 1
print(encoding)
```

```python
position = 0
for i in mutationlist:
    alpha = []
    num = []
    for j in df[i]:
        if j == 0:
            num.append(0)
            alpha.append(0)
        else:
            n = int(str(j)[0:-1])
            print(n)
            num.append(n)
            a = str(j)[-1]
            val = encoding[a]
            alpha.append(val)
    col1 = i+' Num'
    col2 = i+' Alpha'
    position += 1
    df.insert(position,col1,num)
    position +=1
    df.insert(position,col2,alpha)
    position +=1
```

```python
nfvdataset = pd.read_csv("NFVdataset.csv")
nfvdataset
```

```python
nfvinputs = nfvdataset.copy()
del nfvinputs['NFV']
nfvinputs

nfvoutputs = nfvdataset[['NFV']].copy()
nfvoutputs
```

```python
import torch
```

```python
nfvinputs = nfvinputs.to_numpy()
nfvinputs
```

```python
nfvoutputs = nfvoutputs.to_numpy()
nfvoutputs
```

```python
inputs = torch.from_numpy(nfvinputs)
targets = torch.from_numpy(nfvoutputs)
```

```python
print(inputs)
print(targets)
```

```python
from torch.utils.data import TensorDataset
train_ds = TensorDataset(inputs.float(), targets.float())
train_ds[0:3]
```

```python
from torch.utils.data import DataLoader
batch_size = 5
train_dl = DataLoader(train_ds, batch_size, shuffle=True)
```

```python
import torch.nn as nn

model = nn.Linear(14,1)
print(model.weight)
print(model.bias)
```

```python
preds = model(inputs.float())
preds
```

```python
import torch.nn.functional as F
loss_fn = F.mse_loss
loss = loss_fn(model(inputs.float()), targets.float())
print(loss)
```

```python
opt = torch.optim.SGD(model.parameters(), lr=1e-10)
```

```python
# Utility function to train the model
def fit(num_epochs, model, loss_fn, opt, train_dl):
    
    # Repeat for given number of epochs
    for epoch in range(num_epochs):
        
        # Train with batches of data
        for xb,yb in train_dl:
            
            # 1. Generate predictions
            pred = model(xb)
            
            # 2. Calculate loss
            loss = loss_fn(pred, yb)
            
            # 3. Compute gradients
            loss.backward()
            
            # 4. Update parameters using gradients
            opt.step()
            
            # 5. Reset the gradients to zero
            opt.zero_grad()
        
        # Print the progress
        if (epoch+1) % 10 == 0:
            print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```

```python
fit(100, model, loss_fn, opt, train_dl)
```
