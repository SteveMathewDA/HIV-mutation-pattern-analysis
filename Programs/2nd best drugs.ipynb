```python
import math
drugs = ['NFV','SQV','IDV','FPV','ATV','LPV','TPV','DRV']
lst1 = []
lst2 = []
for i in range(len(df)):
    best_Drug1 = ""
    best_Drug2 = ""
    min_resistance1 = 10000
    min_resistance2 = 20000
    for j in drugs:
        a = df._get_value(i,j)
        if a<min_resistance1 and a<min_resistance2:
            min_resistance2 = min_resistance1
            min_resistance1 = a
            best_Drug2 = best_Drug1
            best_Drug1 = j
        if a<min_resistance2 and a>min_resistance1:
            min_resistance2 = a
            best_Drug2 = j
    if best_Drug1!='':
        lst1.append(best_Drug1)
    elif best_Drug2!='':
        lst2.append(best_Drug2)
```

```python
drugs = ['NFV','SQV','IDV','FPV','ATV','LPV','TPV','DRV']
lst = []

for i in range(len(df)):
    temp = []
    druglist = []
    for j in drugs:
        temp.append(df._get_value(i,j))
    temp.sort()
    for k in range(len(temp)):
        for j in drugs:
            if df._get_value(i,j) == temp[k]:
                druglist.append(j)
    lst.append(druglist[1])
        
```
