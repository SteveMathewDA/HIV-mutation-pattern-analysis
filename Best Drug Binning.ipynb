```python
import pandas as pd
```

```python
df = pd.read_csv("datawithnonulls.csv")
```

```python
df
```

```python
raw = df.values.tolist()
```

```python
raw
```

```python
NFV = []
SQV = []
IDV = []
FPV = []
ATV = []
LPV = []
TPV = []
DRV = []

NFV_mc = {}
SQV_mc = {}
IDV_mc = {}
FPV_mc = {}
ATV_mc = {}
LPV_mc = {}
TPV_mc = {}
DRV_mc = {}
```

```python
for i in raw:
    newl = []
    for j in range(7):
        if not pd.isnull(i[j]):
            newl.append(i[j])
    if i[-1] == 'NFV':
        NFV.append(newl)
    elif i[-1] == 'SQV':
        SQV.append(newl)
    elif i[-1] == 'IDV':
        IDV.append(newl)
    elif i[-1] == 'FPV':
        FPV.append(newl)
    elif i[-1] == 'ATV':
        ATV.append(newl)
    elif i[-1] == 'LPV':
        LPV.append(newl)
    elif i[-1] == 'TPV':
        TPV.append(newl)
    elif i[-1] == 'DRV':
        DRV.append(newl)
            
```

```python
len(SQV)
```

```python
for i in SQV:
    for j in i:
        if j not in SQV_mc:
            SQV_mc[j] = 1
        else:
            SQV_mc[j] +=1
            
for i in NFV:
    for j in i:
        if j not in NFV_mc:
            NFV_mc[j] = 1
        else:
            NFV_mc[j] +=1
            
for i in IDV:
    for j in i:
        if j not in IDV_mc:
            IDV_mc[j] = 1
        else:
            IDV_mc[j] +=1
            
for i in FPV:
    for j in i:
        if j not in FPV_mc:
            FPV_mc[j] = 1
        else:
            FPV_mc[j] +=1
            
for i in ATV:
    for j in i:
        if j not in ATV_mc:
            ATV_mc[j] = 1
        else:
            ATV_mc[j] +=1
            
for i in LPV:
    for j in i:
        if j not in LPV_mc:
            LPV_mc[j] = 1
        else:
            LPV_mc[j] +=1
            
for i in TPV:
    for j in i:
        if j not in TPV_mc:
            TPV_mc[j] = 1
        else:
            TPV_mc[j] +=1
            
for i in DRV:
    for j in i:
        if j not in DRV_mc:
            DRV_mc[j] = 1
        else:
            DRV_mc[j] +=1
```
