```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("MappedMatrix.csv")
```

```python
df
```

```python
raw = pd.read_csv("CleanedHIVDataset.csv")
raw
```

```python
nfv_output_wnulls = raw['NFV'].copy()
nfv_output_wnulls = pd.DataFrame(nfv_output_wnulls)
nfv_output_wonulls=nfv_output_wnulls.dropna()
nfv_output_wonulls
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, nfv_output_wonulls,test_size=0.9,random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train,test_size=0.1,random_state=42)
```

```python
model = Sequential()
model.add(Dense(150,input_dim = 36,activation="relu"))
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 
model.add(Dense(150,activation="relu")) 


model.add(Dense(1,activation = "linear"))
model.compile(optimizer='adam',loss="mean_absolute_error", metrics=["mean_absolute_error"])
model.fit(x_train.values,y_train.values, validation_data= (x_val,y_val),epochs=10,batch_size=64)
```

```python
result = model.evaluate(x_test.values,y_test.values)
for i in range(len(model.metrics_names)):
    print("Metric ",model.metrics_names[i],":",str(round(result [i],2)))
```
